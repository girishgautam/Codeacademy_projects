7. Create a new column named genre for each of the games.

    Group the games into their genres: Multiplayer Online Battle Arena (MOBA), First Person Shooter (FPS), Survival, and Other.

    Using CASE, your logic should be:

    If League of Legends → MOBA
    If Dota 2 → MOBA
    If Heroes of the Strom → MOBA
    If Counter-Strike: Global Offensive → FPS
    If DayZ → Survival
    If Survival Evolved → Survival
    Else → Other

```python
# country_group = stream.groupby(['country'])
# country_group['game'].value_counts().loc['US']
```

```python
def fnc (string):
    if string == 'League of Legends':
        return 'MOBA'
    elif string == 'Dota 2':
        return 'MOBA'
    elif string == 'Heroes of the Storm':
        return 'MOBA'
    elif string == 'Counter-Strike: Global Offensive':
        return 'FPS'
    elif string == 'DayZ':
        return 'Survival'
    elif string == 'ARK: Survival Evolved':
        return 'Survival'
    else:
        return 'Other'
```

```python
# game_group = stream.groupby(['game'])
# game_group.get_group('League of Legends')[['country', 'game']]
```

```python
stream['genre'] = stream['game'].apply(fnc)
```

```python
stream.groupby(['game', 'genre'])['game'].count().reset_index(name='count').sort_values('count', ascending=False)
```

8. Let’s write a query that returns three columns:

    The hours of the time column
    The view count for each hour
    Lastly, filter the result with only the users in your country 

```python
stream['time'] =  pd.to_datetime(stream['time'])
```

```python
stream['time'].head()
```

```python
stream['Hour_of_the_day'] = stream['time'].dt.hour
stream.head()
```

```python
filt = stream['country']=='US'
stream.loc[filt].groupby(['Hour_of_the_day'])['Hour_of_the_day'].count().reset_index(name='count').sort_values(by='Hour_of_the_day', ascending=True)
```
